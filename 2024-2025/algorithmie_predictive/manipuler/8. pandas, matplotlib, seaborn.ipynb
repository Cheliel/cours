## Ouverture d'un fichier CSV

```python
filename = "Automobile.csv"
donnes = pd.read_csv(filename, sep=',')
```

## Informations sur les données
* Dimensions des données
* Liste des colonnes
* Premières lignes
* Premières statistiques sur les données

```python
donnes.get('mpg')
donnes.head(60) # les 60 première ligne
donnes.tail(20) # les 20 dernière ligne
donnes.shape() 
donnes.columns()
donnes.describe() 
```

## Accéder aux informations 
* Visualiser une colonne
* Visualiser une ligne
* Visualiser une sous matrice
* Visualiser des lignes pour une colonne donnée

```python
donnes['horsepower'] # get a column 
donnes[['horsepower', 'mpg']] # get columns 

donnes.iloc[1] # get specifique ligne 
donnes.loc[[4, 8, 11]] # get specifiques lignes
donnes['horsepower'][5:10] # get specifiques lign in column

donnes.iloc[5:13, 1:4] # get specific lignes and columns
donnes[['horsepower', 'mpg']][5:10] # get specific columns and lignes

```

## Supprimer des lignes 

```python
donnes.drop([1, 2, 3], axis=0) # Axis 0 -> ligne // Axis 1 -> column
donnes.drop(['mpg', 'name'], axis=1)
donnes # you might want to assign result 
```

## Supprimer des colonnes

```python
toDelet = donnes[donnes['origin'] == 'usa'].index
donnes.drop(toDelet)
```

## Sélection des données
* Sélectionner à partir d'un critère
* Récupérer les index des lignes sélectionnées
* Supprimer les lignes sélectionnées

```python
mask = donnes['mpg'] > 30   # 
mask                        # get boolean value of witch ligne match the predicate
donnes[donnes['mpg'] > 30]  # apply predicate on donnes
donnes[mask]
```

## Quelques valeurs statistiques 
* Compter les valeurs
* Moyenne avec aggrégation des données
* Maximum/Minimum avec aggrégation des données

```python
donnes['mpg'].value_counts()
donnes.groupby(['origin'])['mpg'].mean() # aggragate by origin -> get 'mpg' column -> get mean
donnes.groupby(['origin'])['mpg'].max()
donnes.groupby(['origin'])['mpg'].min()
donnes.groupby(['origin'])['mpg'].describe()['count'] # get a column of the result
donnes.groupby(['origin'])['mpg'].describe()

donnes[donnes['mpg'] == donnes['mpg'].min()] # the ligne with de minimal 'mpg' value
# donnes[donnes['mpg'] == 9.0] # equal 
```
