# Réduction de données (PCA)

## Récupération du modèle 
* On récupère une fois de plus notre modèle d'iris
* Problème : Je ne peux pas représenter graphiquement un modèle composé de 4 paramètres et 1 cible

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris.data
```

## Réduction du modèle
* Grâce à l'algorithme PCA, nous allons ramener les paramètres à leurs composantes les plus significatives en terme de variabilité
* Nous conservons uniquement les 2 composantes principales 
* Le modèle PCA est d'abord créé avec les paramètres requis
* La transformation permet d'obtenir le nombre de composantes voulues

```python
from sklearn.decomposition import PCA
import pylab as pl 

# PCA(n_components=2) => init 
# pca.transform => effectue la transformation des données 

# pour les algo non supervise -> pas de test = tt les donnes sont données
# tt les modèles de SK prennent les même arguments -> data 
# il faut donc simplement répéter les argument de la méthode
# dans le fit on donne les données de train 

# n_components -> utilise les règles mathématique pour déterminer les params important. 
# e.g. s'il y a correlation entre deux colonnes -> suppr. une des deux colonnes (redondance)

pca = PCA(n_components=2).fit(iris.data) 
pca_2d = pca.transform(iris.data)
pca_2d

```

## Visualisation graphique 
* Nous n'avons plus que 2 paramètres, il est maintenant possible de les représenter sur un graphique en 2D
* Nous allons représenter chaque iris en fonction de son type (valeur cible)

```python
for i in range(0, pca_2d.shape[0]):
    if iris.target[i] == 0:
        c1 = pl.scatter(pca_2d[i, 0], pca_2d[i, 1], c='r', marker='+')
    elif iris.target[i] == 1:
        c2 = pl.scatter(pca_2d[i, 0], pca_2d[i, 1], c='g', marker='o')
    elif iris.target[i] == 2:
        c3 = pl.scatter(pca_2d[i, 0], pca_2d[i, 1], c='b', marker='*')


pl.legend([c1, c2, c3], ['Setosia', 'Versicolor', 'Virginica'])
pl.title('Classification de Iris en 3 catégorie connus')
pl.show()

```
